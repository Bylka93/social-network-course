```python
# Finding the max cliques in the graph
G = nx.read_gpickle("Our_Gragh_0.4Sim")
```

```python
cliqs = list(nx.find_cliques(G)) # All cliques
cliqs = sorted(cliqs, key=len, reverse=True) # Sort from largest to smallest
```

```python
len(cliqs) # How many cliques found
```

```python
# Checking clique sizes from largest to smallest
for c in cliqs:
    print(len(c))
```

```python
clique_1st_graph = G.subgraph(cliqs[0]) # Largest clique
clique_2nd_graph = G.subgraph(cliqs[1]) # Second largest
clique_3rd_graph = G.subgraph(cliqs[2]) # Third largest
```

```python
plt.rcParams['figure.figsize'] = (10, 10)
nx.draw(clique_1st_graph, with_labels=True, width=0.5, node_color='lightgreen', edge_color='grey')
plt.savefig("largest_clique.png")
```

```python
nx.draw(clique_2nd_graph, with_labels=True, width=0.5, node_color='lightgreen', edge_color='grey')
plt.savefig("second_largest_clique.png")
```

```python
nx.draw(clique_3rd_graph, with_labels=True, width=0.5, node_color='lightgreen', edge_color='grey')
plt.savefig("third_largest_clique.png")
```

```python
# Finding the betweenness centrality from the graph

G = nx.read_gpickle("Our_Gragh_0.4Sim")

Gg2 = nx.edge_betweenness_centrality(G, normalized=True, weight=None)
nx.draw(Gg2, with_labels=True)
plt.show()
```

```python
# Finding the k-core from the graph

G = nx.read_gpickle("Our_Gragh_0.4Sim")
Gg3 = nx.k_core(G)
nx.draw(Gg3, with_labels=True)
plt.show()
```
