```python
# Finding the max cliques in the graph
G = nx.read_gpickle("Our_Gragh_0.4Sim")

Gg = nx.make_max_clique_graph(G, create_using=None)

nx.draw(Gg, with_labels=True)
plt.show()
```

```python
# Finding the betweenness centrality from the graph

G = nx.read_gpickle("Our_Gragh_0.4Sim")

Gg = nx.edge_betweenness_centrality(G, normalized=True, weight=None)
nx.draw(Gg, with_labels=True)
plt.show()
```

```python
# Finding the k-core from the graph

G = nx.read_gpickle("Our_Gragh_0.4Sim")
Gg = nx.k_core(G)
nx.draw(Gg, with_labels=True)
plt.show()
```
