## Top 10 influencers in each hashtag in terms of number of tweets generated

```python
import pandas as pd
import numpy as np

parking = pd.read_csv('parking_from_january.csv')
parkinglot = pd.read_csv('parkinglot_from_january.csv')
parkinggarage = pd.read_csv('parkinggarage_from_january.csv')
```

## \#Parking

```python
top10parking = parking.groupby(['user_id','username']).count().sort_values('time', ascending=False)['time'].iloc[:10]
data_parking = pd.DataFrame({'id': top10parking.index.get_level_values(0), 'name': top10parking.index.get_level_values(1), 'number_of_tweets': top10parking.values})
data_parking.to_csv('top10influencers_parking.csv', index = False)
```

## \#ParkingLot

```python
top10parkinglot = parkinglot.groupby(['user_id','username']).count().sort_values('time', ascending=False)['time'].iloc[:10]
data_parkinglot = pd.DataFrame({'id': top10parkinglot.index.get_level_values(0), 'name': top10parkinglot.index.get_level_values(1), 'number_of_tweets': top10parkinglot.values})
data_parkinglot.to_csv('top10influencers_parkinglot.csv', index = False)
```

## \#ParkingGarage

```python
top10parkinggarage = parkinggarage.groupby(['user_id','username']).count().sort_values('time', ascending=False)['time'].iloc[:10]
data_parkinggarage = pd.DataFrame({'id': top10parkinggarage.index.get_level_values(0), 'name': top10parkinggarage.index.get_level_values(1), 'number_of_tweets': top10parkinggarage.values})
data_parkinggarage.to_csv('top10influencers_parkinggarage.csv', index = False)
```
